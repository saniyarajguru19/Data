```python
import pandas as pd
import numpy as np
df=pd.DataFrame({'Rollno':[1,2,np.nan,4,5],
                   'Name':[np.nan,'Saniya','Sakshi','Prithvi','Bhumika'],
                   'Surname':['Rajguru',np.nan,'Shetty','Shetty','Phadke'],
                   'Age':[19,18,np.nan,19,18],
                   'Marks':[60,np.nan,50,80,40],
                   'Class':["A","B","A",np.nan,"B"]})
df
```

```python
df.isnull().sum()
```

```python
df["Rollno"]=df["Rollno"].fillna(3)
df["Rollno"]
```

```python
mean_age=df["Age"].mean()
mean_age
```

```python
df["Marks"]=df["Marks"].fillna(mean_age)
df["Marks"]
```

```python
df
```

```python
median_age=df["Age"].median()
median_age
```

```python
mode_age=df["Age"].mode()
mode_age
```
